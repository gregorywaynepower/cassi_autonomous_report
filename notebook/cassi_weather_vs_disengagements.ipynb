```python
#collects weather data 
weather_df = pd.read_csv('../data/cassi_weather.csv', parse_dates=True, index_col='time')
weather_df.drop(columns=['snow', 'wpgt', 'tsun'], inplace=True)
#formats columns as floats and category for weather condition
weather_df.coco = weather_df.coco.astype("category")
weather_df.rename(columns={'temp':'Temperature_Celsius','rhum':'Relative_Humidity_%','dwpt':'Dew_Point','prcp':'Precipitation','wdir':'Wind_Direction_Degrees','wspd':'Windspeed_mph','pres':'Air_Pressure_hPa','coco':'Weather_Condition'},inplace=True)
weather_df
```

```python
#collects shuttle data
shuttle_df = pd.read_csv('../data/autonomous-shuttle-disengagement.csv', parse_dates=True, index_col='Incident Datetime')
# Add UTC Parsing so Timestamps are Correct
shuttle_df.index= pd.to_datetime(shuttle_df.index, utc=True)
#formats timestamp by rounding to nearest hour and removing offset
shuttle_df['time'] = [time.round('60min').to_pydatetime().replace(tzinfo=None) for time in shuttle_df.index]
shuttle_df.set_index('time',inplace=True, drop=True)

#merges weather and shuttle data   
df = pd.merge(shuttle_df, weather_df, left_index=True, right_index=True)

#print final dataframe
df
```
